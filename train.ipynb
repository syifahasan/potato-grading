```python
import os
from ultralytics import YOLO
```

```python
BASE_DIR = os.path.dirname(os.path.abspath(__file__))
DATA_YAML = os.path.join(BASE_DIR, 'dataset', 'data.yaml')

model = YOLO('yolo26n-seg.pt')
results = model.train(
    data=DATA_YAML,
    epochs=100,
    batch=8,
    imgsz=640,
    amp=True,
    device=0
)
```
